```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\satya\Downloads\Superstore.csv", encoding='ISO-8859-1')
df.head()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], errors = 'coerce')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], errors = 'coerce')
df.head()
```

```python
print(df.isnull().sum())  # Show nulls per column
df = df.dropna()
df.head()
```

```python
sns.set(style='whitegrid')
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
# Total sales Over Time by using matplotlib 
sales_time = df.groupby('Year-Month')['Sales'].sum().reset_index()
sns.lineplot(x='Year-Month', y='Sales', data=sales_time, marker='o')
plt.title('Total Sales Over Time')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Sales by category ("Category vs Total Sales")
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False).reset_index()
sns.barplot(x='Sales', y='Category', data=category_sales, palette='Blues_d')
plt.title('Sales by Category')
plt.xlabel('Total Sales')
plt.ylabel('Category')
plt.tight_layout()
plt.show()
```

```python
# Profit by Sub-Category (Sub-Category vs Profit)
subcategory_profit = df.groupby('Sub-Category')['Profit'].sum().sort_values().reset_index()
sns.barplot(x='Profit', y='Sub-Category', data=subcategory_profit, palette='coolwarm')
plt.title('Profit by Sub-Category')
plt.tight_layout()
plt.show()
```

```python
# Sales vs Profit bt Segement
segment_df = df.groupby('Segment')[['Sales', 'Profit']].sum().reset_index()
plt.scatter(segment_df['Sales'], segment_df['Profit'], s=segment_df['Sales']/100, alpha=0.6)
for i in range(segment_df.shape[0]):
    plt.text(segment_df['Sales'][i], segment_df['Profit'][i], segment_df['Segment'][i])
plt.title('Sales vs Profit by Segment')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.tight_layout()
plt.show()
```

```python
# Sales Heatmap ("Region vs Category")
pivot_table = pd.pivot_table(df, values='Sales', index='Region', columns='Category', aggfunc='sum')
sns.heatmap(pivot_table, annot=True, fmt=".0f", cmap='YlGnBu')
plt.title('Sales Heatmap: Region vs Category')
plt.tight_layout()
plt.show()
```
